# KL-Divergence vs Cross Entropy


```python
import sys
from matplotlib import pyplot as plt
import numpy as np
import math
from scipy.stats import norm
```

## Defining distributions

```python
x = np.arange(1,10,0.01)

p = norm.pdf(x , 5 , 1 )
q = norm.pdf(x , 6 , 1 )
```

```python
fig, ax = plt.subplots(facecolor="w",figsize=(10,10))
plt.style.use('fivethirtyeight')
ax.plot(x,p)
ax.set_xticks([])
ax.set_yticklabels([])
ax.grid(False)


ax.fill_between(x,p,0, alpha=0.3, color='r')
ax.plot(x,q)
ax.fill_between(x,q,0, alpha=0.2)

ax.text(5, 0.3, r'$\mathcal{P}$', fontsize=12)
ax.text(6, 0.3, r'$\mathcal{Q}$', fontsize=12)

ax.text(0.7, 0.3, r'$\mathcal{D}_{KL}(\mathcal{P}||\mathcal{Q})$ = $Entropy(\mathcal{P},\mathcal{Q})$?', fontsize=16)
```

# KL-Divergence

![kl-divergence](images/kl.png)

```python
def D_kl(p,q):
    return sum(p*np.log2(p) - p*np.log2(q))
```

```python
D_kl(p,q)
```

## Entropy

![entropy](images/entropy_p.png)

```python
def entropy(p):
    return -1* sum(p*np.log2(p))
```

```python
entropy(p)
```

## Cross Entropy

![entropy](images/entropy_pq.png)

```python
def cross_entropy(p,q):
    return -1* sum(p*np.log2(q))
```

```python
cross_entropy(p,q)
```

## Cross Entropy = KL Divergence?

```python
cross_entropy(p,q) == D_kl(p,q) + entropy(p)
```
